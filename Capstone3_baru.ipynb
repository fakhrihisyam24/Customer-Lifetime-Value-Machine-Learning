# 1 | Pendahuluan

Customer Lifetime Value (CLV) adalah metrik penting yang mengukur nilai total yang diharapkan dari seorang pelanggan selama masa hubungan mereka dengan perusahaan. Memahami CLV memungkinkan perusahaan untuk mengidentifikasi pelanggan yang paling berharga dan merencanakan strategi pemasaran yang lebih efektif. Dengan informasi ini, perusahaan dapat mengoptimalkan pengeluaran untuk akuisisi pelanggan dan mempertahankan pelanggan yang ada, sehingga meningkatkan profitabilitas jangka panjang.

Pada penelitian oleh Gupta et al. (2006) dalam studi berjudul "Valuing Customers" menunjukkan bahwa biaya untuk mendapatkan pelanggan baru bisa 5 hingga 25 kali lebih tinggi dibandingkan dengan biaya untuk mempertahankan pelanggan yang sudah ada. Hal ini menekankan pentingnya fokus pada retensi pelanggan, karena meningkatkan retensi pelanggan sebesar 5% dapat meningkatkan keuntungan antara 25% hingga 95%, tergantung pada industri. Dengan menganalisis CLV, perusahaan dapat menyesuaikan penawaran produk dan layanan sesuai dengan kebutuhan dan preferensi pelanggan, yang pada gilirannya dapat meningkatkan kepuasan dan loyalitas. Oleh karena itu, pemahaman yang mendalam tentang CLV menjadi sangat penting bagi perusahaan yang ingin mencapai pertumbuhan yang berkelanjutan dan sukses di pasar.

Selain itu, perusahaan yang menerapkan strategi berbasis CLV sering kali dapat mengidentifikasi segmen pelanggan yang paling menguntungkan dan mengalokasikan sumber daya mereka dengan lebih efisien. Dengan menggunakan data analitik untuk memahami perilaku dan preferensi pelanggan, perusahaan dapat merancang kampanye pemasaran yang lebih tepat sasaran dan meningkatkan pengalaman pelanggan secara keseluruhan. Ini tidak hanya membantu dalam meningkatkan retensi, tetapi juga mendorong pelanggan untuk melakukan pembelian berulang, yang pada akhirnya berkontribusi pada pertumbuhan pendapatan dan profitabilitas jangka panjang. Dengan demikian, CLV bukan hanya sekadar angka, tetapi merupakan alat strategis yang dapat mengubah cara perusahaan berinteraksi dengan pelanggan mereka.

## 1.1 | Customer Lifetime Value

Customer Lifetime Value (CLV) adalah metrik yang digunakan untuk memperkirakan total nilai yang dapat dihasilkan oleh seorang pelanggan selama masa hubungan mereka dengan perusahaan. CLV memberikan wawasan tentang seberapa berharga seorang pelanggan bagi bisnis, tidak hanya dari segi pendapatan yang dihasilkan, tetapi juga dari segi biaya yang dikeluarkan untuk memperoleh dan mempertahankan pelanggan tersebut.

Komponen Utama CLV
1. Pendapatan yang Dihasilkan: Ini mencakup semua pendapatan yang dihasilkan dari pelanggan selama masa hubungan mereka dengan perusahaan. Dalam konteks asuransi mobil, ini bisa berupa premi yang dibayarkan oleh pelanggan setiap tahun.
2. Biaya Akuisisi Pelanggan (CAC): Ini adalah biaya yang dikeluarkan untuk mendapatkan pelanggan baru, termasuk biaya pemasaran, iklan, dan promosi. Memahami CAC penting untuk menghitung CLV secara akurat.
3. Biaya Pemeliharaan Pelanggan: Ini mencakup biaya yang dikeluarkan untuk mempertahankan pelanggan, seperti layanan pelanggan, dukungan teknis, dan program loyalitas.
4. Durasi Hubungan Pelanggan: Ini adalah periode waktu di mana pelanggan diperkirakan akan tetap berhubungan dengan perusahaan. Durasi ini dapat bervariasi tergantung pada jenis produk atau layanan yang ditawarkan.

## 1.2 | Manfaat Customer Lifetime Value (CLV)
Menghitung dan memahami Customer Lifetime Value (CLV) memberikan berbagai manfaat bagi perusahaan, terutama dalam konteks strategi pemasaran dan pengelolaan hubungan pelanggan. Berikut adalah beberapa manfaat utama dari adanya CLV:
1. **Pengambilan Keputusan yang Lebih Baik**:
   - CLV memberikan wawasan yang mendalam tentang nilai pelanggan, memungkinkan perusahaan untuk membuat keputusan yang lebih terinformasi terkait investasi dalam pemasaran dan akuisisi pelanggan.

2. **Optimalisasi Strategi Pemasaran**:
   - Dengan mengetahui pelanggan mana yang memiliki CLV tinggi, perusahaan dapat merancang kampanye pemasaran yang lebih terarah dan efektif, serta mengalokasikan anggaran pemasaran dengan lebih efisien.

3. **Peningkatan Retensi Pelanggan**:
   - Memahami CLV membantu perusahaan dalam mengidentifikasi faktor-faktor yang mempengaruhi kepuasan dan loyalitas pelanggan, sehingga dapat mengembangkan program loyalitas dan inisiatif retensi yang lebih baik.

4. **Peningkatan Profitabilitas**:
   - Dengan fokus pada pelanggan yang memiliki CLV tinggi, perusahaan dapat meningkatkan profitabilitas mereka. Mengurangi biaya akuisisi dan meningkatkan retensi pelanggan dapat menghasilkan peningkatan pendapatan yang signifikan.

5. **Segmentasi Pelanggan yang Lebih Efektif**:
   - CLV memungkinkan perusahaan untuk melakukan segmentasi pelanggan berdasarkan nilai yang dihasilkan, membantu dalam merancang penawaran yang lebih relevan dan menarik bagi setiap segmen.

Dengan memahami dan memanfaatkan manfaat-manfaat ini, perusahaan dapat meningkatkan kinerja bisnis secara keseluruhan dan membangun hubungan yang lebih kuat dengan pelanggan mereka.

## 1.5 | Metriks Evaluasi

1. **RMSE** (Root Mean Squared Error)

RMSE adalah akar kuadrat dari rata-rata kesalahan kuadrat antara nilai yang diprediksi dan nilai aktual. RMSE memberikan bobot lebih pada kesalahan yang lebih besar, sehingga sangat sensitif terhadap outlier. Nilai RMSE yang lebih rendah menunjukkan model yang lebih baik. RMSE diukur dalam satuan yang sama dengan variabel target, sehingga mudah untuk diinterpretasikan.

![Gambar Contoh](RMSE.jpg)  
```

2. **MAE** (Mean Absolute Error)

MAE adalah rata-rata dari kesalahan absolut antara nilai yang diprediksi dan nilai aktual.  MAE memberikan gambaran yang lebih sederhana tentang kesalahan rata-rata tanpa memberikan bobot lebih pada kesalahan yang lebih besar. MAE juga diukur dalam satuan yang sama dengan variabel target. Nilai MAE yang lebih rendah menunjukkan model yang lebih baik, dan MAE lebih robust terhadap outlier dibandingkan RMSE.

```

MAPE adalah rata-rata dari kesalahan absolut yang dinyatakan dalam persentase dari nilai aktual. MAPE memberikan informasi tentang kesalahan dalam konteks persentase, sehingga lebih mudah untuk diinterpretasikan dalam konteks bisnis. Nilai MAPE yang lebih rendah menunjukkan model yang lebih baik. Namun, MAPE tidak dapat dihitung jika ada nilai aktual yang sama dengan nol, dan dapat memberikan hasil yang menyesatkan jika nilai aktual bervariasi secara signifikan.

### 2.11 | Multikolinearitas dengan VIF

````python
# buat function
def calc_vif(x):
    # dataframe kosong
    vif = pd.DataFrame()
    vif['variables'] = x.columns
    # menghitung vig untuk setiap kolom nya
    vif['VIF'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
    return vif
````

````python
# Rentang yang ditentukan  
ranges = [(1000, 10000), (10000, 20000), (20000, 30000), (30000, 40000),  
          (40000, 50000), (50000, 60000), (60000, 70000), (70000, 80000),  
          (80000, 90000)]  
  
# DataFrame untuk menyimpan hasil  
results = []  
  
# Hitung MAPE untuk setiap rentang  
for lower_bound, upper_bound in ranges:  
    # Filter data berdasarkan rentang  
    filtered_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
    filtered_df = filtered_df[(filtered_df['Actual'] >= lower_bound) & (filtered_df['Actual'] < upper_bound)]  
      
    # Hitung MAPE  
    if not filtered_df.empty:  
        mape = (abs(filtered_df['Actual'] - filtered_df['Predicted']) / filtered_df['Actual']).mean() * 100  
        results.append({'Range': f'{lower_bound} - {upper_bound}', 'MAPE': f'{mape:.2f}%'})  
    else:  
        results.append({'Range': f'{lower_bound} - {upper_bound}', 'MAPE': 'N/A'})  
  
# Buat DataFrame untuk hasil  
results_df = pd.DataFrame(results)  
  
# Tampilkan hasil  
results_df
````

Dapat dilihat untuk prediksi dari range 1.000 - 10.000, MAPE hanya sebesar 5.85% saja. Yang artinya, semakin kecil CLVnya, semakin kecil juga presentase errornya.

* Akurasi Model: Model menunjukkan akurasi yang baik di rentang 1,000 - 10,000, tetapi akurasi menurun secara signifikan di rentang yang lebih tinggi.

* Perbaikan Diperlukan: MAPE yang tinggi di rentang 20,000 ke atas menunjukkan bahwa model perlu ditingkatkan, baik melalui pemilihan fitur yang lebih baik, tuning parameter, atau menggunakan algoritma yang lebih kuat.

* Evaluasi Lebih Lanjut: Penting untuk melakukan analisis lebih lanjut untuk memahami mengapa model tidak berkinerja baik di rentang yang lebih tinggi dan untuk mencari solusi yang tepat.

* Asumsi  
  
- **Pendapatan Bulanan per Pelanggan**: $150    
- **Biaya Promosi per Pelanggan**: $10    
- **Biaya Akuisisi Pelanggan Baru**: $100    
- **MAPE**: 11.8% (Mean Absolute Percentage Error)    
- **Jumlah Pelanggan**: 1000 (Total pelanggan yang diprediksi)    
  
* Perhitungan Keuntungan dan Kerugian  
  
Total Pendapatan yang Diharapkan  
$$  
\text{Total Pendapatan} = \text{Jumlah Pelanggan} \times \text{Pendapatan Bulanan per Pelanggan} = 1000 \times 150 = 150,000  
$$  
  
Estimasi Kerugian Berdasarkan MAPE  
$$  
\text{Estimasi Kerugian} = \text{Total Pendapatan} \times \left(\frac{\text{MAPE}}{100}\right) = 150,000 \times 0.118 = 17,700  
$$  
  
Total Biaya Pemeliharaan  
- **Total Biaya Promosi**:    
$$  
\text{Total Biaya Promosi} = \text{Jumlah Pelanggan} \times \text{Biaya Promosi per Pelanggan} = 1000 \times 10 = 10,000  
$$  
  
- **Total Biaya Akuisisi**:    
$$  
\text{Total Biaya Akuisisi} = \text{Jumlah Pelanggan} \times \text{Biaya Akuisisi per Pelanggan} = 1000 \times 100 = 100,000  
$$  
  
Total Biaya  
$$  
\text{Total Biaya} = \text{Total Biaya Promosi} + \text{Total Biaya Akuisisi} = 10,000 + 100,000 = 110,000  
$$  
  
Keuntungan Bersih  
$$  
\text{Keuntungan Bersih} = \text{Total Pendapatan} - \text{Total Biaya} - \text{Estimasi Kerugian} = 150,000 - 110,000 - 17,700 = 22,300  
$$  
  
Output Hasil  
- **Total Pendapatan yang Diharapkan**: $150,000.00    
- **Estimasi Kerugian karena MAPE**: $17,700.00    
- **Total Biaya Promosi**: $10,000.00    
- **Total Biaya Akuisisi**: $100,000.00    
- **Total Biaya**: $110,000.00    
- **Keuntungan Bersih**: $22,300.00
